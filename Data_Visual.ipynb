```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file into a pandas DataFrame
df = pd.read_csv('user_data.csv')

# Show the ages with the highest income
highest_income = df[df['total_income'] == df['total_income'].max()]
plt.figure(figsize=(10, 6))
plt.hist(highest_income['age'], bins=20, color='skyblue', edgecolor='black')
plt.title('Distribution of Ages with Highest Income')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.grid(True)

# Save the chart as an image file
plt.savefig('highest_income_chart.png')

# Show the gender distribution across spending categories
spending_categories = ['utilities', 'entertainment', 'school_fees', 'shopping', 'healthcare']
gender_spending = df.groupby('gender')[spending_categories].sum()

plt.figure(figsize=(10, 6))
gender_spending.plot(kind='bar', stacked=True)
plt.title('Gender Distribution Across Spending Categories')
plt.xlabel('Gender')
plt.ylabel('Total Spending')
plt.xticks(rotation=0)
plt.legend(title='Spending Category')

# Save the chart as an image file
plt.savefig('gender_spending_chart.png')

# Display the charts
plt.show()
```
